### **5.2 Stop count list**

```python
stop_count[:2]
```

This table was not splitted when read. So it must be splitted, first after each ' " ' 

**5.2.1 Splitting stop count table**

```python
stop_count_expand = stop_count.iloc[:,0].str.split(pat='"', expand=True) # 
stop_count_expand[:2]
```

Some columns contain only commas or are empty at all. These can be deleted in a next step.

**5.2.2 Deleting unneccessary columns**

```python
stop_count_expand.drop([2, 4, 8, 9, 10, 12, 14, 16, 19, 20, 22], axis=1, inplace=True) # dropping comma and empty columns
```

**5.2.3 Renaming the columns**

```python
stop_count_expand.rename(columns={0: "year", 1: "comp_Nr.", 3: "comp_name", 5: "comp_short", 6: "id_numbers", 7:"stat_name", 11: "canton", 13: "SLOID", 15: "tr_type", 17: "nr_days", 18: "schedule_time", 21: "end_st", 23: "line", 24:"BP_ID"}, inplace=True)
stop_count_expand[:2]
```

Now we see that there are many columns which must be cleaned in a next step. Some contains unneccessary commas, other must be splitted again.

**5.2.4 Cleaning all columns**

```python
# column "year": contains unneccessary comma
stop_count_expand["year"] = stop_count_expand["year"].str.split(pat=',', expand=True)[0] # remove ","
```

```python
# column "id_numbers": Contains two values, ride id and stop id, needs to be separated
stop_count_expand["ride_id"] = stop_count_expand["id_numbers"].str.split(pat=',', expand=True)[[1]] # first number is ride identification number
stop_count_expand["stop_id"] = stop_count_expand["id_numbers"].str.split(pat=',', expand=True)[[2]] # second number is stop identification number
stop_count_expand.drop("id_numbers", axis=1, inplace=True) # remove id_numbers column which is not needed anymore
```

```python
# column "schedule_time": Contains Departure and arrival time, needs to be separated
stop_count_expand["arr_time"] = stop_count_expand["schedule_time"].str.split(pat=',', expand=True)[[1]] # first number is arrival
stop_count_expand["dep_time"] = stop_count_expand["schedule_time"].str.split(pat=',', expand=True)[[2]] # second number is departure
stop_count_expand.drop("schedule_time", axis=1, inplace=True) # remove id_numbers column which is not needed anymore

# only time is needed
stop_count_expand["arr_time"] = pd.to_datetime(stop_count_expand["arr_time"])
stop_count_expand["dep_time"] = pd.to_datetime(stop_count_expand["dep_time"])
```

```python
# column "BP_ID": contains leading comma
stop_count_expand["BP_ID"] = stop_count_expand["BP_ID"].str.split(pat=',', expand=True)[0] # remove ","
```

Now the table seems to be ok and can be written into a csv.
